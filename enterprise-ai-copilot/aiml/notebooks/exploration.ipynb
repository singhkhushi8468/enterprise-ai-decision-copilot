```python
score = score_decision({
    'domain': 'Market Expansion',
    'risk': 'Moderate',
    'horizon': 12,
    'confidence': 72
})
print(f"Confidence: {score['overall_confidence']}% (Grade: {score['grade']})")
print(f"Recommendation: {score['recommendation']}")
print('Factors:', score['factors'])
```

## 3. Confidence Interval

```python
ci = calculate_confidence_interval(value=34.0, confidence=72.0, n=30)
print(f"Revenue uplift: {ci['estimate']}% [{ci['lower']}%, {ci['upper']}%]")
```

## 4. Risk Matrix

```python
risks = [
    {'name': 'EU Regulatory', 'probability': 0.7, 'impact': 0.8, 'category': 'Regulatory'},
    {'name': 'Cloud dependency', 'probability': 0.4, 'impact': 0.6, 'category': 'Operational'},
    {'name': 'Competitor entry', 'probability': 0.5, 'impact': 0.5, 'category': 'Market'},
]
matrix = compute_risk_matrix(risks)
print(f"Overall risk: {matrix['overall_score']} ({matrix['overall_level']})")
for r in matrix['risks']:
    print(f"  {r['name']}: {r['score']} ({r['level']})")
```

## 5. SHAP-style Explanation

```python
contribs = explain_prediction(
    feature_values={'market_size': 0.85, 'competition': 0.42, 'regulatory': 0.65, 'data_quality': 0.94},
    feature_weights={'market_size': 0.35, 'competition': -0.25, 'regulatory': -0.20, 'data_quality': 0.30},
    prediction=34.0
)
for c in contribs:
    arrow = '+' if c['direction'] == 'positive' else '-'
    print(f"  [{arrow}] {c['feature']}: {c['contribution']:+.2f} (importance {c['importance_pct']}%)")
```
